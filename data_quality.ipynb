```python
import pandas as pd
import os, glob
```

```python
data_quality = pd.DataFrame({"file": [], "attribute": [], "completeness": [], "uniqueness": []})
data_quality
```

```python
na_values = ["", 
             "#N/A", 
             "#N/A N/A", 
             "#NA", 
             "-1.#IND", 
             "-1.#QNAN", 
             "-NaN", 
             "-nan", 
             "1.#IND", 
             "1.#QNAN", 
             "<NA>", 
             "N/A", 
#              "NA", 
             "NULL", 
             "NaN", 
             "n/a", 
             "nan", 
             "null"]
```

```python
for file in glob.glob(os.curdir + "/sources/*.csv"):
    df = pd.read_csv(file, sep=';', na_values=na_values, keep_default_na=False)
    
    missing_percentage = df.isnull().mean() * 100
    completeness = 100 - missing_percentage
    
    if (file == './sources\\pages_listing.csv'):
        uniqueness = df.duplicated(subset='Page URL').sum() / df.count() * 100
    elif (file == './sources\\browser_list.csv'):
        uniqueness = df.duplicated(subset='Browser code').sum() / df.count() * 100
    elif (file == './sources\\country_listing.csv'):
        uniqueness = df.duplicated(subset='Country code').sum() / df.count() * 100
    elif (file == './sources\\OS_listing.csv'):
        uniqueness = df.duplicated(subset='OS code').sum() / df.count() * 100
    elif (file == './sources\\pages_topics.csv'):
        uniqueness = df.duplicated(subset='Topic code').sum() / df.count() * 100
    elif (file == './sources\\session_listing.csv'):
        uniqueness = df.duplicated(subset='Session Id').sum() / df.count() * 100
    else:
        uniqueness = df.duplicated().sum() / df.count() * 100
    
    uniqueness = 100 - uniqueness

   # consistency = df.value_counts(normalize=True) / df.count() * 100
   # print(consistency)

    new_rows = pd.DataFrame({'file': file, "attribute": completeness.index, "completeness": completeness.values, "uniqueness": uniqueness.values})
    data_quality = pd.concat([data_quality, new_rows])
```

```python
data_quality_file = data_quality.filter(['file', 'completeness', 'uniqueness']).groupby('file', as_index=False).mean()
data_quality_file
```
